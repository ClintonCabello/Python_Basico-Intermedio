## **Datos Outliers**

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
edades = np.array([4,17,29,50,51,61,71,85,92,99])
len(edades)
```

```python
plt.hist(edades, bins = 5, color = 'green', ec = 'black')
plt.title('Edades')
plt.show()
```

```python
plt.boxplot(edades)
plt.xlabel('Edades')
plt.grid(True)
plt.show()
```

```python
print(f'Min: {np.min(edades)}')
print(f'Q1: {np.quantile(edades,0.25)}')
print(f'Med: {np.median(edades)}')
print(f'Q3: {np.quantile(edades,0.75)}')
print(f'Max: {np.max(edades)}')
```

## Regresión Lineal

```python
import pandas as pd
data = pd.DataFrame({'AÑO(X)':[2015,2016,2017,2018,2019,2020],
                    'CIRUGIAS(Y)':[100,124,142,165,158,174]})
data
```

```python
plt.scatter(data['AÑO(X)'], data['CIRUGIAS(Y)'], color = 'red')
plt.xlabel('AÑO')
plt.ylabel('CIRUGIAS')
plt.show()
```

```python
from sklearn import linear_model
regresion = linear_model.LinearRegression()
año = data['AÑO(X)'].values.reshape(-1,1)
modelo = regresion.fit(año, data['CIRUGIAS(Y)'])
print('Intercepto',modelo.intercept_)
print('Pendiente',modelo.coef_)
```

```python
# predecir para el año 2023
predecir =[[2023]]
modelo.predict(predecir)
```
