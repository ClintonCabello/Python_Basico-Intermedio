```python
import numpy as np
a√±os = ['2010','2011','2012','2013','2014','2015','2016','2017']
nivel1 = np.random.rand(8)*100
nivel2 = np.random.rand(8)*200+100
nivel3 = np.random.rand(8)*300+200
nivel4 = np.random.rand(8)*400+300
nivel5 = np.random.rand(8)*500+400
```

```python
plt.plot(a√±os,nivel1, label = "Nivel 1", color = 'black', marker='.')
plt.plot(a√±os,nivel2, label = 'Nivel 2', linestyle = '--')
plt.plot(a√±os,nivel3, label = 'Nivel 3')
plt.plot(a√±os,nivel4, label = 'Nivel 4')
plt.plot(a√±os,nivel5, label = 'Nivel 5', marker = '>')
plt.legend(loc = 'upper right')
plt.title('Examen de Certificaci√≥n')
plt.xlabel('A√±os en que se ha realizado el examen')
plt.ylabel('Notas')
plt.yticks(np.arange(0,1000,100))
plt.show()
```
